## List（列表）
List（列表） 是 Python 中使用最频繁的数据类型。

列表可以完成大多数集合类的数据结构实现。列表中元素的类型可以不相同，它支持数字，字符串甚至可以包含列表（所谓嵌套）。

### 索引与分片
一图流：


```
List[start : end : step]
```
参数：
- start: 表示切片开始的位置（包括该位置）
- end: 表示切片结束的位置（不包括该位置）
- step: 表示切片的步长,负数表示逆序输出，默认步长为1


````python
list = [ 'gentle', 114514, 1919810 ]
print(list) # 完整列表
print(list[0]) # 列表第一个元素
````

````python
print (list[1:3]) # 从第二个开始输出到第三个元素
print (list[2:]) # 输出从第三个元素开始的所有元素
list[::-1]
````

### 列表的其他函数/方法
```
len(list) # 列表元素个数

list(tuple) # 将元组转换为列表

list.append(obj) # 在列表末尾添加新的对象

list.extend(seq) # 在列表末尾一次性追加另一个序列中的多个值（用新列表扩展原来的列表）
list.index(obj) # 从列表中找出某个值第一个匹配项的索引位置
list.insert(index, obj) # 将对象插入列表
list.pop([index=-1]) # 移除列表中的一个元素（默认最后一个元素），并且返回该元素的值
list.remove(obj) # 移除列表中某个值的第一个匹配项
list.reverse() # 反向列表中元素
list.sort(key=None, reverse=False) # 对原列表进行排序
list.clear() # 清空列表
list.copy() 复制列表

```

split()方法的语法：

## 循环

### while循环
```
while 判断条件(condition)：
    执行语句(statements)
    ...
```

````python
n = 100
 
sum = 0
counter = 1
while counter <= n:
    sum = sum + counter
    counter += 1
 
sum
````

````python
# while -- else

count = 0
while count < 5:
   print (count, " 小于 5")
   count = count + 1
else:
   print (count, " 大于或等于 5")
````

### range函数
语法：
```

```
参数：
- start: 计数从 start 开始。默认是从 0 开始
- stop: 计数到 stop 结束，但不包括 stop

#### Tips
map() 会根据提供的函数对指定序列做映射。

语法：
```
map(function, iterable, ...)
```
参数：
- function: 函数
- iterable: 一个或多个序列

具体详见：https://www.runoob.com/python/python-func-map.html

````python
n = int(input())
a = input().split(' ',n)
a.pop()
a = list(map(int, a)) # 防止输入过多
a
````

### 例题
将下面C/C++代码翻译成python

来自：https://codeforces.com/gym/102028/problem/I
```cpp
#include <iostream>
#define maxn 100010
#define ll long long 
using namespace std;

int n, a[maxn], d[maxn];

ll s[maxn];

void work() {
    cin >> n; d[1] = 0; 
    for (int i = 2; i <= n; ++i) {
        cin >> a[i];
        d[i] = d[i - 1] + a[i];
    }
    for (int i = 1; i <= n; ++i) s[i] = s[i - 1] + d[i]; 
    cout << 0 << " " << d[n];
    ll ans = d[n]; 
    for (int l = 1, r = n; l < r - 1;) {
        ans += 1ll * l * d[l + 1] - s[l];
        ans += (s[n] - s[r - 1]) - 1ll * (n - r + 1) * d[l + 1];
        ++l;
        cout << " " << ans; 
        if (l >= r - 1) break; 
        ans += 1ll * l * d[r - 1] - s[l];
        ans += (s[n] - s[r - 1]) - 1ll * (n - r + 1) * d[r - 1];
        --r;
        cout << " " << ans; 
    } cout << "\n"; 
} 

int main() { 
    ios::sync_with_stdio(false);
    cin.tie(nullptr); cout.tie(nullptr);

    int T; cin >> T;
    while (T--) work(); 
    return 0;
} 
```
